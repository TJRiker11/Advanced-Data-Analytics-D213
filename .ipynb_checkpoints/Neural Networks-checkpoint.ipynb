# Neural Networks

```python
import json
import pandas as pd
import gzip
```

```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)
```

```python
def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
```

```python
df = getDF('Review Data/Musical_Instruments_5.json.gz')
```
